```python
# Unitless Mileage
# 1.40 conversion rate via Google search
unitlessMileage = []
for iterator in df["Mileage"]:
    if str(iterator).endswith("km/kg"):
        iterator = iterator[:-6]
        iterator = float(iterator) * 1.40
        unitlessMileage.append(float(iterator))
    elif str(iterator).endswith("kmpl"):
        iterator = iterator[:-6]
        unitlessMileage.append(float(iterator))
    else:
        str(iterator) == np.nan
        iterator = np.nan
        unitlessMileage.append(iterator)
```

```python
# Reassign df['Mileage'] such that it represents unitlessMileage
df["Mileage"] = unitlessMileage
```

```python
# While doing univariate analysis of numerical variables we want to study their central tendency and dispersion.
# This is a function that will create a boxplot and a histogram for any input numerical variable.
# This function takes the numerical column as the input and returns the boxplots and histograms for the variable.


def histogram_boxplot2(feature, figsize=(15, 10), bins=None):
    """Boxplot and histogram combined
    feature: 1-d feature array
    figsize: size of fig (default (9,8))
    bins: number of bins (default None / auto)
    """
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows=2,  # Number of rows of the subplot grid= 2
        sharex=True,  # x-axis will be shared among all subplots
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize,
    )  # creating the 2 subplots
    sns.boxplot(
        feature, ax=ax_box2, showmeans=True, color="violet"
    )  # boxplot will be created and a star will indicate the mean value of the column
    sns.distplot(
        feature, kde=F, ax=ax_hist2, bins=bins, palette="winter"
    ) if bins else sns.distplot(
        feature, kde=False, ax=ax_hist2
    )  # For histogram
    ax_hist2.axvline(
        feature.mean(), color="green", linestyle="--"
    )  # Add mean to the histogram
    ax_hist2.axvline(
        feature.median(), color="black", linestyle="-"
    )  # Add median to the histogram
```

```python
# CREATE A BARPLOT FUNCTION (countplot) THAT INDICATES THE COUNT FOR EACH CATEGORY


def category_count_barplot(data, column):
    """
    plot
    feature: categorical feature
    the function will not work if a column is passed as a hue parameter
    """
    # need the total length of the column to use as a denominator in a subsequent equation
    total_length = len(df[column])

    plt.figure(figsize=(15, 5))
    plt.xticks(rotation=45)
    ax = sns.countplot(data=df, x=column, palette="Paired")
    for p in ax.patches:
        # get the count for each class of the category
        percentage = p.get_height()
        # get the width of the plot
        x = p.get_x() + p.get_width() / total_length + 0.2
        # get the heigth of the plot
        y = p.get_y() + p.get_height()

        # annotate the percentages
        ax.annotate(percentage, (x, y), size=12)
    # show the plot
    plt.show()
```

* The model explains 64% and 62.8% of the total variation in the train and test sets respectively. 
* Both scores seem to be comparable.

    
    
RMSE values on the train and test sets are also comparable: training_set=6.235 and testing_set=6.166 (Oooops!). 

    
    
The model is overfitting.  Lots of noise.   Lots of rookie coding errors.

    
    
MAE indicates that the current model is able to predict use car pricing within a mean error of 320% on the test set (Oooops!). 

```python
# Adjusted R^2
def adj_r2(ind_vars, targets, predictions):
    r2 = r2_score(targets, predictions)
    n = ind_vars.shape[0]
    k = ind_vars.shape[1]
    return 1 - ((1 - r2) * (n - 1) / (n - k - 1))


# Model performance check
def model_perf(model, inp, out):

    y_pred = model.predict(inp)
    y_act = out.values

    return pd.DataFrame(
        {
            "RMSE": np.sqrt(mean_squared_error(y_act, y_pred)),
            "MAE": mean_absolute_error(y_act, y_pred),
            "R^2": r2_score(y_act, y_pred),
            "Adjusted R^2": adj_r2(inp, y_act, y_pred),
        },
        index=[0],
    )
```

## 45. INTERACTION TERMS
    
    
<font color=#0e92ea>
   
    
    
* By including non-linear terms, interaction terms attempts to improve upon the R^2 testing score
    
    
* Import polynomial features (degree=2 --> X^2 or X1*X2; but utilize interaction_only=True (no squares X^2)
    

* By adding transformations to each of the old sets (X_train and X_test), this process redefine both training data set (X_train_2) and testing data set (X_test_2) with newer (and more) columns
    
    
* Polynomial Features (with only interaction terms) have improved the testing (out-of-sample) R^2. 
    
    
* Note the significant increase in the number of variables just to improve the R^2 testing score

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

poly = PolynomialFeatures(degree=2, interaction_only=True)
X_train2 = poly.fit_transform(X_train)
X_test2 = poly.fit_transform(X_test)

# A new linear regression model
poly_clf = linear_model.LinearRegression()
# Fit the model with the newer training data columns (such that they include the polynomial interaction terms)
poly_clf.fit(X_train2, y_train)

y_pred = poly_clf.predict(X_test2)

print(y_pred)
```

```python
# In-sample (training) fit: R^2 will always improve as the number of variables increases
print(poly_clf.score(X_train2, y_train))
```

```python
# Out-of-sample (testing) fit: R^2 does not improve!
print(poly_clf.score(X_test2, y_test))
```

## 46. FORWARD FEATURE SELECTION
    
    
<font color=#0e92ea>
   
    
    
* Forward feature selection begins with an empty model and adds variables one-by-one.

    
* In each forward step, the function sequentially adds the variable that gives the single best improvement to the model.

    
* Use forward feature selection on all the variables.


```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

reg = LinearRegression()

# Build step forward feature selection
sfs = SFS(
    reg,
    k_features=X_train.shape[1],
    forward=True,  # k_features denotes "Number of features to select"
    floating=False,
    scoring="r2",
    n_jobs=-1,
    verbose=2,
    cv=5,
)

# Perform SFS.fit()
sfs = sfs.fit(X_train, y_train)
```

```python
# Plot the performance with addition of each feature
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs

fig1 = plot_sfs(sfs.get_metric_dict(), kind="std_err", figsize=(15, 5))
plt.title("Sequential Forward Selection (w. StdErr)")
plt.xticks(rotation=90)
plt.show()
```

```python
reg = LinearRegression()

# # Build step forward feature selection
sfs = SFS(
    reg,
    k_features=30,
    forward=True,
    floating=False,
    scoring="r2",
    n_jobs=-1,
    verbose=2,
    cv=5,
)

# Perform SFS.fit()
sfs = sfs.fit(X_train, y_train)
```
